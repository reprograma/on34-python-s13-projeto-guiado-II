```python
# https://github.com/maria-gomes23/on34-python-s13-projeto-guiado-II.git

import pandas as pd

df = pd.read_csv('/content/banklist.csv', sep=',', encoding='latin-1', on_bad_lines='skip')

df.columns = df.columns.str.strip()

df = df.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remover espaços extras dos nomes das colunas

df['Ano de Fechamento'] = pd.to_datetime(df['Closing Date']).dt.year  # Extrair o ano

plt.figure(figsize=(10, 6))
sns.lineplot(x='Ano de Fechamento', y='Fund', data=df)
plt.title('Gráfico de Linha da Série Temporal')
plt.xlabel('Ano de Fechamento')
plt.ylabel('Fundos')
plt.show()

```
